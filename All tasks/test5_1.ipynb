```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.rc('font', **{'size' : 22})
```

```python
x = np.array([0.0,1.0,2.0,3.0,4.0])
y = np.array([4.0,2.5,1.0,-1.0,-2.0])
```

```python
def l(x,a,k):
    res = 1.0
    for i in range(len(x)):
        if i !=k :
            res = res*(a-x[i])/(x[k]-x[i])
    return res
def interpolation(x,y,a): # x,y -точки по которым строится функция лагранжа , a - точка где хотим посчитать ее значение 
    res =[]
    for j in range(len(a)):
        res.append(0.0)
        for i in range(len(x)):
            res[j] = res[j] + y[i]*l(x,a[j],i)
    return np.array(res)
def l_diff(x,a,k):
    res = 0.0
    for i in range(len(x)):
        if i !=k :
            temp = 1.0
            for j in range(len(x)):
                if j != i and j != k :
                    temp = temp*(a-x[j])
            res = res + temp
    for i in range(len(x)):
        if i != k :
            res = res/(x[k]-x[i])
    return res
def diff_interpolation(x,y,a):
    res =[]
    for j in range(len(a)):
        res.append(0.0)
        for i in range(len(x)):
            res[j] = res[j] + y[i]*l_diff(x,a[j],i)
    return np.array(res)
```

```python
x_list = np.arange(0.0,4.0,0.01)
y_list = interpolation(x,y,x_list)
plt.figure(figsize=(10, 10))
plt.plot(x,y , 'k.', ms=15, label='Заданная функция по 4 точкам')
plt.plot(x_list,y_list , 'k-', ms=15, label='интерполяция функции')
print("Производная через интерполяцию :",diff_interpolation(x,y,[3.0])[0])
coef = np.array([1.0,-2.0,1.0])
print(-coef@y[2:])
print(y[:-1])
plt.legend()
plt.show()
```
