```python

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.rc('font', **{'size' : 22})

ax = plt.figure(figsize=(18, 10)).add_subplot(projection='3d')

# Prepare arrays x, y, z
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)

ax.plot(x, y, z,'k--', ms=16, label='')

plt.xlabel('$x$')
plt.ylabel('y')
plt.legend()
plt.show()
```
