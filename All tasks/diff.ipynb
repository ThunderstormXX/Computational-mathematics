```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.rc('font', **{'size' : 22})
```

```python
def f(x):
    return x*x*x

def diff2_left(array,h):
    result = []
    for i in range(len(array)-3):
        result.append( 
            (2*f(array[i]) -5*f(array[i+1]) +4*f(array[i+2]) -f(array[i+3]) )/(h*h)
        )
    return result

def diff2_middle(array,h):
    result = []
    for i in range(1,len(array)-1):
        result.append( 
            (f(array[i+1]) -2*f(array[i]) +f(array[i-1])  )/(h*h)
        )
    return result
def diff2_right(array,h):
    result = []
    for i in range(3,len(array)):
        result.append( 
            (2*f(array[i-3]) -5*f(array[i-2]) +4*f(array[i-1]) -f(array[i]) )/(h*h)        
         )
    return result
```

```python
h = 0.00001
h_array = np.arange(0.0,6.0,h)

array_solve = diff2_right(h_array , h)
h_array = h_array[3:]

# array_solve = diff2_left(h_array , h)
# h_array = h_array[:-3]

# array_solve = diff2_middle(h_array , h)
# h_array = h_array[1:-1]


# array_solve = psi(h_array)

plt.figure(figsize=(18, 10))
plt.plot(h_array , array_solve , 'k-', ms=16, label='Решение')
plt.legend()
plt.show()
```
