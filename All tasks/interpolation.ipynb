```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.rc('font', **{'size' : 22})
```

```python
x = np.array([0,np.pi/6 , np.pi/4 ,np.pi/3])
y = np.array([0,0.5,0.71 ,0.87])
```

```python
def f(x):
    return 4*x - np.cos(x)
def l(x,a,k):
    res = 1.0
    for i in range(len(x)):
        if i !=k :
            res = res*(a-x[i])/(x[k]-x[i])
    return res
def interpolation(x,y,a): # x,y -точки по которым строится функция лагранжа , a - точка где хотим посчитать ее значение 
    res =[]
    for j in range(len(a)):
        res.append(0.0)
        for i in range(len(x)):
            res[j] = res[j] + y[i]*l(x,a[j],i)
    return np.array(res)
def w_abs(x,a): 
    res =[]
    for j in range(len(a)):
        res.append(0.0)
        for i in range(len(x)):
            res[j] = res[j] + np.abs(l(x,a[j],i))
    return np.array(res)
```

```python
h = 0.01
z = np.arange(0.0,1.0,h)
phi = [7*np.pi/24.0]
y_2 = interpolation(x,y,z)

plt.figure(figsize=(10, 10))
plt.plot(x,y , 'k.', ms=15, label='Заданная функция по 4 точкам')
plt.plot(z, y_2, 'k--', ms=15, label='Ее интерполяция')

print(w_abs(x,phi)[0])

plt.legend()
plt.show()
```
